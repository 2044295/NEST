# Goal

- Translate hex-code color input into neural signal
- Integrate/fire a red/green/blue neuron based on color

### Setup

```python
import nest
import numpy as np
import matplotlib.pyplot as plt
```

### Translation

```python
def get_num(num, x): return( (num & int('FF', 16) << 8 * x) >> 8 * x )
def get_bits(num): return(np.intp(np.array(list('{0:0>8b}'.format(num)))))


color = int('f5f5f5', 16)

red = get_num(color, 2)
green = get_num(color, 1)
blue = get_num(color, 0)

red_bits = get_bits(red)
green_bits = get_bits(green)
blue_bits = get_bits(blue)

bits = [red, green, blue]

# print(red_bits, green_bits, blue_bits)
max_bits = []
if red == max(bits): max_bits.append('Red')
if green == max(bits): max_bits.append('Green')
if blue == max(bits): max_bits.append('Blue')

if len(max_bits) > 1:
    print('Max Colors: {}'.format(', '.join(max_bits)))
else:
    print('Max Color: {}'.format(max_bits[0]))

print(red, green, blue)
print(red_bits, green_bits, blue_bits)
```

### Transduction + Integration + Decision

- Weight for \~ 1:1 - 1200.0
- Inhibitory values are (excitatory values - 1) to make equal values not cancel

```python
nest.ResetKernel()

red_params = [{'I_e': x} for x in 376.0 * red_bits]
green_params = [{'I_e': x} for x in 376.0 * green_bits]
blue_params = [{'I_e': x} for x in 376.0 * blue_bits]

red_neurons = nest.Create("iaf_psc_alpha", 8, params=red_params)
green_neurons = nest.Create("iaf_psc_alpha", 8, params=green_params)
blue_neurons = nest.Create("iaf_psc_alpha", 8, params=blue_params)
pre_multimeter = nest.Create("multimeter", params={"withtime":True, "record_from":["V_m"]})


all_neurons = red_neurons + green_neurons + blue_neurons
nest.Connect(pre_multimeter, all_neurons)

""""""

values = np.float64(2 ** np.arange(0, 8)[::-1])

syn_spec = [[{'weight': x} for x in 1200 * values], [{'weight': x} for x in -1200 * values]]
# print(syn_spec_ex, syn_spec_in)

red_green = nest.Create("iaf_psc_alpha", 2, params={'I_e': 375.9, 't_ref': 20.0})  #red >= green, green >= red
green_blue = nest.Create("iaf_psc_alpha", 2, params={'I_e': 375.9, 't_ref': 20.0}) #green >= blue, blue >= green
blue_red = nest.Create("iaf_psc_alpha", 2, params={'I_e': 375.9, 't_ref': 20.0})   #blue >= red, red >= blue
trans_detector = nest.Create("spike_detector", params={"withgid":True, "withtime":True})
trans_multimeter = nest.Create("multimeter", params={"withtime":True, "record_from":["V_m"]})

all_trans = red_green + green_blue + blue_red
nest.Connect(all_trans, trans_detector)
nest.Connect(trans_multimeter, all_trans)

for x in range(len(red_green)):
    for y in range(len(red_neurons)):
        nest.Connect([red_neurons[y]], [red_green[x]], syn_spec=syn_spec[x][y])
        nest.Connect([green_neurons[y]], [red_green[x]], syn_spec=syn_spec[1-x][y])
        nest.Connect([green_neurons[y]], [green_blue[x]], syn_spec=syn_spec[x][y])
        nest.Connect([blue_neurons[y]], [green_blue[x]], syn_spec=syn_spec[1-x][y])
        nest.Connect([blue_neurons[y]], [blue_red[x]], syn_spec=syn_spec[x][y])
        nest.Connect([red_neurons[y]], [blue_red[x]], syn_spec=syn_spec[1-x][y])

""""""

red_final = nest.Create("iaf_psc_alpha")
green_final = nest.Create("iaf_psc_alpha")
blue_final = nest.Create("iaf_psc_alpha")
final_detector = nest.Create("spike_detector", params={"withgid":True, "withtime":True})
final_multimeter = nest.Create("multimeter", params={"withtime":True, "record_from":["V_m"]})

all_finals = red_final + green_final + blue_final
nest.Connect(all_finals, final_detector)
nest.Connect(final_multimeter, all_finals)

nest.Connect([red_green[0]], red_final, syn_spec={'weight': 600.0})
nest.Connect([blue_red[1]], red_final, syn_spec={'weight': 600.0})

nest.Connect([green_blue[0]], green_final, syn_spec={'weight': 600.0})
nest.Connect([red_green[1]], green_final, syn_spec={'weight': 600.0})

nest.Connect([blue_red[0]], blue_final, syn_spec={'weight': 600.0})
nest.Connect([green_blue[1]], blue_final, syn_spec={'weight': 600.0})

""""""

nest.Simulate(100.0)

"""
""""""

status = nest.GetStatus(pre_multimeter)[0]
senders = status['events']['senders']
V_m = status['events']['V_m']
time = status['events']['times']

shape = (99, 3, 8)
senders = senders.reshape(shape)
V_m = V_m.reshape(shape)
time = time.reshape(shape)

fig, axes = plt.subplots(8, 3, figsize=(8, 16))

for group, axe in enumerate(axes):
    for num, ax in enumerate(axe):
        #print(group, num), print(senders[:,num,group])
        ax.plot(time[:,num,group], V_m[:,num,group])

plt.show(block=False)

""""""
"""

status = nest.GetStatus(trans_detector)[0]
senders = status['events']['senders']
time = status['events']['times']

fig, axes = plt.subplots(1, 6, figsize=(16, 4))

for x in range(len(all_trans)):
    times = time[np.where(senders == all_trans[x])]
    axes[x].plot(times, [all_trans[x]] * len(times), 'bo')

plt.show(block=False)

""""""

status = nest.GetStatus(trans_multimeter)[0]
senders = status['events']['senders']
V_m = status['events']['V_m']
time = status['events']['times']

shape = (99, 6)
senders = senders.reshape(shape)
V_m = V_m.reshape(shape)
time = time.reshape(shape)

fig, axes = plt.subplots(1, 6, figsize=(16, 4))

for num, ax in enumerate(axes):
    ax.plot(time[:,num], V_m[:,num])

plt.show(block=False)

""""""

status = nest.GetStatus(final_detector)[0]
senders = status['events']['senders']
time = status['events']['times']

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for x in range(len(all_finals)):
    times = time[np.where(senders == all_finals[x])]
    axes[x].plot(times, [all_finals[x]] * len(times), 'bo')

plt.show(block=False)

""""""

status = nest.GetStatus(final_multimeter)[0]
senders = status['events']['senders']
V_m = status['events']['V_m']
time = status['events']['times']

shape = (99, 3)
senders = senders.reshape(shape)
V_m = V_m.reshape(shape)
time = time.reshape(shape)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

for num, ax in enumerate(axes):
    ax.plot(time[:,num], V_m[:,num])

plt.show(block=False)
```
